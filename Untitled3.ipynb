```python
# Real-time Jupyter Dashboard for Deribit Premium & Funding Analysis
# Updates graphs after each data collection

import requests
import time
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
from collections import deque
import warnings
warnings.filterwarnings('ignore')

# For real-time updates in Jupyter
from IPython.display import display, clear_output
import matplotlib.dates as mdates
from matplotlib.animation import FuncAnimation

# Set style
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

class RealTimeDeribitDashboard:
    def __init__(self, instrument="BTC-PERPETUAL", history_length=500):
        self.instrument = instrument
        self.currency = instrument.split('-')[0]
        self.deribit_base_url = "https://www.deribit.com/api/v2/public"
        
        # Data storage
        self.history_length = history_length
        self.data_history = deque(maxlen=history_length)
        
        # Dashboard parameters
        self.update_count = 0
        self.start_time = datetime.now()
        
        # Analysis parameters
        self.volume_window = 20
        self.pressure_threshold = 0.3
        self.base_interest_rate = 0.0001
        self.funding_cap = 0.0005
        
        # Initialize plots
        self.fig = None
        self.axes = None
        
        print(f"ðŸš€ Real-time Dashboard initialized for {instrument}")
        
    def fetch_ticker_data(self):
        """Fetch basic ticker data"""
        try:
            url = f"{self.deribit_base_url}/ticker"
            params = {"instrument_name": self.instrument}
            response = requests.get(url, params=params, timeout=5)
            
            if response.status_code == 200:
                return response.json()["result"]
            return None
        except Exception as e:
            return None
    
    def fetch_index_price(self):
        """Fetch index price"""
        try:
            index_name = f"{self.currency.lower()}_usd"
            url = f"{self.deribit_base_url}/get_index_price"
            params = {"index_name": index_name}
            response = requests.get(url, params=params, timeout=5)
            
            if response.status_code == 200:
                return response.json()["result"]["index_price"]
            return None
        except Exception as e:
            return None
    
    def fetch_order_book(self, depth=10):
        """Fetch order book data"""
        try:
            url = f"{self.deribit_base_url}/get_order_book"
            params = {
                "instrument_name": self.instrument,
                "depth": depth
            }
            response = requests.get(url, params=params, timeout=5)
            
            if response.status_code == 200:
                result = response.json()["result"]
                return {
                    "bids": result["bids"],
                    "asks": result["asks"],
                    "timestamp": result["timestamp"]
                }
            return None
        except Exception as e:
            return None
    
    def fetch_recent_trades(self, count=50):
        """Fetch recent trades"""
        try:
            url = f"{self.deribit_base_url}/get_last_trades_by_instrument"
            params = {
                "instrument_name": self.instrument,
                "count": count
            }
            response = requests.get(url, params=params, timeout=5)
            
            if response.status_code == 200:
                return response.json()["result"]["trades"]
            return None
        except Exception as e:
            return None
    
    def analyze_market_pressure(self, order_book, trades):
        """Analyze market pressure"""
        if not order_book:
            return {}
        
        try:
            # Order book analysis
            bid_volume = sum([float(bid[1]) for bid in order_book["bids"]])
            ask_volume = sum([float(ask[1]) for ask in order_book["asks"]])
            total_volume = bid_volume + ask_volume
            
            bid_ratio = bid_volume / total_volume if total_volume > 0 else 0.5
            ask_ratio = ask_volume / total_volume if total_volume > 0 else 0.5
            pressure_score = bid_ratio - ask_ratio
            
            # Trade analysis
            trade_pressure = 0
            if trades:
                buy_trades = [t for t in trades if t["direction"] == "buy"]
                sell_trades = [t for t in trades if t["direction"] == "sell"]
                
                buy_volume = sum([float(t["amount"]) for t in buy_trades])
                sell_volume = sum([float(t["amount"]) for t in sell_trades])
                total_trade_volume = buy_volume + sell_volume
                
                if total_trade_volume > 0:
                    trade_pressure = (buy_volume - sell_volume) / total_trade_volume
            
            return {
                "bid_volume": bid_volume,
                "ask_volume": ask_volume,
                "pressure_score": pressure_score,
                "trade_pressure": trade_pressure,
                "volume_imbalance": (bid_volume - ask_volume) / total_volume if total_volume > 0 else 0
            }
        except Exception as e:
            return {}
    
    def predict_funding_rate(self, premium_index, pressure_score, current_funding):
        """Simple funding prediction"""
        interest_component = self.base_interest_rate
        premium_component = premium_index
        pressure_adjustment = pressure_score * 0.1
        
        predicted_funding = interest_component + premium_component + pressure_adjustment
        predicted_funding = max(-self.funding_cap, min(self.funding_cap, predicted_funding))
        
        return {
            "predicted_funding": predicted_funding,
            "annual_rate": predicted_funding * 365 * 3,
            "daily_rate": predicted_funding * 3
        }
    
    def collect_market_data(self):
        """Collect all market data"""
        timestamp = datetime.now()
        
        ticker = self.fetch_ticker_data()
        index_price = self.fetch_index_price()
        order_book = self.fetch_order_book()
        trades = self.fetch_recent_trades()
        
        if not all([ticker, index_price]):
            return None
        
        mark_price = ticker["mark_price"]
        funding_8h = ticker["funding_8h"]
        premium_rate = ((mark_price - index_price) / index_price) * 100
        premium_index = premium_rate / 100
        
        pressure_data = self.analyze_market_pressure(order_book, trades)
        funding_pred = self.predict_funding_rate(premium_index, pressure_data.get('pressure_score', 0), funding_8h)
        
        # Risk assessment
        risk_level = "LOW_RISK"
        if premium_rate > 0.1 and pressure_data.get('pressure_score', 0) > 0.3 and funding_8h > 0.01:
            risk_level = "HIGH_RISK"
        elif premium_rate > 0.05 and pressure_data.get('pressure_score', 0) > 0.2:
            risk_level = "MEDIUM_RISK"
        
        data_point = {
            "timestamp": timestamp,
            "mark_price": mark_price,
            "index_price": index_price,
            "premium_rate": premium_rate,
            "premium_index": premium_index,
            "funding_8h": funding_8h,
            "funding_annual": funding_8h * 365 * 3,
            "risk_level": risk_level,
            **pressure_data,
            **funding_pred
        }
        
        return data_point
    
    def create_realtime_dashboard(self):
        """Create real-time updating dashboard"""
        # Clear previous output
        clear_output(wait=True)
        
        if len(self.data_history) < 2:
            print("ðŸ“Š Collecting initial data...")
            return
        
        # Convert to DataFrame
        df = pd.DataFrame(list(self.data_history))
        df.set_index('timestamp', inplace=True)
        
        # Calculate moving averages
        if len(df) >= 5:
            df['premium_ma'] = df['premium_rate'].rolling(5).mean()
            df['pressure_ma'] = df['pressure_score'].rolling(5).mean()
        
        # Create subplots
        fig, axes = plt.subplots(2, 2, figsize=(16, 12))
        fig.suptitle(f'{self.instrument} - Real-time Dashboard (Update #{self.update_count})', 
                     fontsize=16, fontweight='bold')
        
        # Plot 1: Premium vs Funding
        ax1 = axes[0, 0]
        ax1.plot(df.index, df['premium_rate'], label='Premium Rate %', color='blue', linewidth=2)
        if 'premium_ma' in df.columns:
            ax1.plot(df.index, df['premium_ma'], label='Premium MA(5)', color='lightblue', alpha=0.7)
        
        ax1_twin = ax1.twinx()
        ax1_twin.plot(df.index, df['funding_8h'] * 100, label='Current Funding %', color='red', alpha=0.7)
        ax1_twin.plot(df.index, df['predicted_funding'] * 100, label='Predicted Funding %', 
                     color='orange', linestyle='--', alpha=0.8)
        
        ax1.axhline(y=0, color='black', linestyle='--', alpha=0.5)
        ax1.set_ylabel('Premium Rate %', color='blue')
        ax1_twin.set_ylabel('Funding Rate %', color='red')
        ax1.set_title('Premium vs Funding Rates')
        ax1.legend(loc='upper left')
        ax1_twin.legend(loc='upper right')
        ax1.grid(True, alpha=0.3)
        
        # Plot 2: Market Pressure
        ax2 = axes[0, 1]
        ax2.plot(df.index, df['pressure_score'], label='Order Book Pressure', color='green', linewidth=2)
        ax2.plot(df.index, df['trade_pressure'], label='Trade Pressure', color='orange', alpha=0.8)
        if 'pressure_ma' in df.columns:
            ax2.plot(df.index, df['pressure_ma'], label='Pressure MA(5)', color='darkgreen', alpha=0.6)
        
        ax2.axhline(y=0, color='black', linestyle='--', alpha=0.5)
        ax2.axhline(y=0.3, color='red', linestyle=':', alpha=0.7, label='High Pressure')
        ax2.axhline(y=-0.3, color='red', linestyle=':', alpha=0.7)
        ax2.set_ylabel('Pressure Score')
        ax2.set_title('Market Pressure Analysis')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
        
        # Plot 3: Volume Distribution
        ax3 = axes[1, 0]
        width = (df.index[-1] - df.index[0]).total_seconds() / len(df) / 86400  # Auto width
        ax3.bar(df.index, df['bid_volume'], alpha=0.6, color='green', label='Bid Volume', width=width)
        ax3.bar(df.index, -df['ask_volume'], alpha=0.6, color='red', label='Ask Volume', width=width)
        ax3.set_ylabel('Volume')
        ax3.set_title('Order Book Volume Distribution')
        ax3.legend()
        ax3.grid(True, alpha=0.3)
        
        # Plot 4: Risk & Stats
        ax4 = axes[1, 1]
        
        # Risk level colors
        risk_colors = {'HIGH_RISK': 'red', 'MEDIUM_RISK': 'orange', 'LOW_RISK': 'green'}
        for risk_level, color in risk_colors.items():
            mask = df['risk_level'] == risk_level
            if mask.any():
                ax4.scatter(df[mask].index, df[mask]['premium_rate'], 
                           c=color, label=f'{risk_level}', alpha=0.7, s=50)
        
        ax4.plot(df.index, df['premium_rate'], alpha=0.3, color='gray', linewidth=1)
        ax4.axhline(y=0, color='black', linestyle='--', alpha=0.5)
        ax4.set_ylabel('Premium Rate %')
        ax4.set_title('Risk Assessment Over Time')
        ax4.legend()
        ax4.grid(True, alpha=0.3)
        
        # Format x-axis for all plots
        for ax in axes.flat:
            ax.xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))
            ax.xaxis.set_major_locator(mdates.MinuteLocator(interval=max(1, len(df)//10)))
            plt.setp(ax.xaxis.get_majorticklabels(), rotation=45)
        
        plt.tight_layout()
        plt.show()
        
        # Print latest stats
        latest = df.iloc[-1]
        runtime = datetime.now() - self.start_time
        
        print(f"\nðŸ“Š LATEST STATS (Runtime: {runtime})")
        print("=" * 60)
        print(f"ðŸ’° Premium: {latest['premium_rate']:+.4f}% | Funding: {latest['funding_8h']:+.4f}%")
        print(f"ðŸ”® Predicted Funding: {latest['predicted_funding']:+.4f}% (Annual: {latest['annual_rate']:+.1f}%)")
        print(f"ðŸ”„ Pressure: OB={latest['pressure_score']:+.3f}, Trade={latest['trade_pressure']:+.3f}")
        print(f"âš ï¸ Risk Level: {latest['risk_level']}")
        print(f"ðŸ“¦ Volume: Bid={latest['bid_volume']:,.0f}, Ask={latest['ask_volume']:,.0f}")
        print("=" * 60)
    
    def run_realtime_monitoring(self, updates=50, interval=30):
        """Run real-time monitoring with live dashboard updates"""
        print(f"ðŸš€ Starting real-time monitoring...")
        print(f"ðŸ“Š {updates} updates every {interval} seconds")
        print(f"ðŸ“ˆ Dashboard will update after each data collection")
        print("â¹ï¸ Press Kernel > Interrupt to stop")
        
        try:
            for i in range(updates):
                self.update_count = i + 1
                
                # Collect data
                data = self.collect_market_data()
                if data:
                    self.data_history.append(data)
                
                # Update dashboard
                self.create_realtime_dashboard()
                
                # Wait for next update
                if i < updates - 1:
                    print(f"â³ Next update in {interval} seconds... (Update {i+1}/{updates})")
                    time.sleep(interval)
                
        except KeyboardInterrupt:
            print(f"\nâ¹ï¸ Monitoring stopped by user after {self.update_count} updates")
        except Exception as e:
            print(f"âŒ Error during monitoring: {e}")
        
        print(f"\nâœ… Real-time monitoring completed!")
        print(f"ðŸ“Š Total updates: {self.update_count}")
        print(f"ðŸ“ˆ Final dashboard shows last {len(self.data_history)} data points")

# Quick functions for easy use
def start_realtime_dashboard(updates=30, interval=60):
    """Start real-time dashboard"""
    dashboard = RealTimeDeribitDashboard()
    dashboard.run_realtime_monitoring(updates=updates, interval=interval)

def quick_realtime_demo():
    """Quick 10-update demo"""
    dashboard = RealTimeDeribitDashboard()
    dashboard.run_realtime_monitoring(updates=100, interval=10)

if __name__ == "__main__":
    print("ðŸ“Š Real-time Deribit Dashboard")
    print("=" * 50)
    print("Available functions:")
    print("- start_realtime_dashboard() - Full monitoring")
    print("- quick_realtime_demo() - 10-update demo")
    print("=" * 50)
    print("ðŸ’¡ Tip: Charts update automatically after each data collection!")
```
